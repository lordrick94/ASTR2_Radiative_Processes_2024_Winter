```python
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def spherical_to_cartesian(r, theta, phi):
    """
    Convert spherical coordinates to Cartesian coordinates.
    
    Parameters:
    r (float): length of the vector.
    theta (float): azimuthal angle in radians.
    phi (float): polar angle in radians.
    
    Returns:
    tuple: Cartesian coordinates (x, y, z).
    """
    x = r * math.sin(phi) * math.cos(theta)
    y = r * math.sin(phi) * math.sin(theta)
    z = r * math.cos(phi)
    
    return x, y, z

# Example usage
r = 5  # Length of the vector
theta = math.radians(45)  # Convert 45 degrees to radians for azimuthal angle
phi = math.radians(60)  # Convert 60 degrees to radians for polar angle

# Get Cartesian coordinates
x, y, z = spherical_to_cartesian(r, theta, phi)

# Create a new figure for 3D plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the vector
# Note: The quiver function in 3D expects the start point and the components of the vectors
ax.quiver(0, 0, 0, x, y, z, color='b', arrow_length_ratio=0.1)

# Setting the plot limits, adjust as necessary
ax.set_xlim([0, max(x, r)])
ax.set_ylim([0, max(y, r)])
ax.set_zlim([0, max(z, r)])

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.title('3D Vector from Spherical Coordinates')
plt.show()
```
